```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
RND = pd.read_csv('RND.txt')
```

```python
RND
```

```python
RND1 = RND.iloc[::3, :]
```

```python
RND1
```

```python
RND2 = RND.iloc[1::3, :]
```

```python
RND2
```

```python
RND3 = RND.iloc[2::3, :]
```

```python
ax = RND1.plot.scatter(x = "n", y = "swaps", color = "red", label = "k_mid")
RND2.plot.scatter(ax = ax, x = "n", y = "swaps", color = "green", label = "k_beg")
RND3.plot.scatter(ax = ax, x = "n", y = "swaps", color = "blue", label = "k_end")


plt.ylabel("Swaps")
plt.legend(loc = "upper left")
plt.title("RNG Swaps")
plt.savefig('RNG_Swaps.png', bbox_inches='tight')
plt.show()
```

```python
ax = RND1.plot.scatter(x = "n", y = "comparisions", color = "red", label = "k_mid")
RND2.plot.scatter(ax = ax, x = "n", y = "comparisions",  color = "green", label = "k_beg")
RND3.plot.scatter(ax = ax, x = "n", y = "comparisions",  color = "blue", label = "k_end")


plt.ylabel("Comparisions")
plt.legend(loc = "upper left")
plt.title("RNG Comparisions")
plt.savefig('RNG_Comparisions.png', bbox_inches='tight')
plt.show()
```

# BS


```python
BS_low = pd.read_csv('BS_low.txt')
BS_high = pd.read_csv('BS_high.txt')
BS_mid = pd.read_csv('BS_mid.txt')
BS_out = pd.read_csv('BS_out.txt')
BS_rnd = pd.read_csv('BS_rnd.txt')
```

```python
BS_rnd
```

```python
ax = BS_low.plot.scatter(x = "n", y = "comparisions", color = "red", label = "BS_low")
BS_high.plot.scatter(ax = ax, x = "n", y = "comparisions", color = "green", label = "BS_high")
BS_mid.plot.scatter(ax = ax, x = "n", y = "comparisions",s = 3, color = "blue", label = "BS_mid")
BS_out.plot.scatter(ax = ax, x = "n", y = "comparisions", color = "pink", label = "BS_out")
BS_rnd.plot.scatter(ax = ax, x = "n", y = "comparisions",s = 1, color = "black", label = "BS_rnd")

plt.ylabel("comparisions")
plt.legend(loc = "upper left")
plt.title("BS comparisions")
plt.savefig('BS_comparisions.png', bbox_inches='tight')
plt.show()
```

```python
ax = BS_low.plot.scatter(x = "n", y = "duration", color = "red", label = "BS_low")
BS_high.plot.scatter(ax = ax, x = "n", y = "duration", color = "green", label = "BS_high")
BS_mid.plot.scatter(ax = ax, x = "n", y = "duration", color = "blue", label = "BS_mid")
BS_out.plot.scatter(ax = ax, x = "n", y = "duration", color = "pink", label = "BS_out")
BS_rnd.plot.scatter(ax = ax, x = "n", y = "duration",s =1, color = "black", label = "BS_rnd")

plt.ylabel("duration")
plt.legend(loc = "upper left")
plt.title("BS duration")
plt.savefig('BS_duration.png', bbox_inches='tight')
plt.show()
```
